```python
import numpy as np
```

```python
phones = np.load('./phoneList.npy',allow_pickle=True)
phones.shape
```

```python
for phone in phones:
    print("Phone: ", phone)
    X = np.load('./CAMBRIDGE_BIGGER/'+phone+'_X.npy', allow_pickle=True)
    np.random.shuffle(X)
    print("Total X shape: ", X.shape)
    
    X_train = X[:int(np.ceil(X.shape[0]*0.7))]
    X_test = X[int(np.ceil(X.shape[0]*0.7)):int(np.ceil(X.shape[0]*0.9))]
    X_val = X[int(np.ceil(X.shape[0]*0.9)):]
    
    print("Shape of X_train: ", X_train.shape)
    print("Shape of X_val: ", X_val.shape)
    print("Shape of X_test: ", X_test.shape)

    np.save("./Cambridge_bigger Phones Train/"+phone+"_X.npy",X_train)
    np.save("./Cambridge_bigger Phones Dev/"+phone+"_X.npy",X_val)
    np.save("./Cambridge_bigger Phones Test/"+phone+"_X.npy",X_test)
    
    print("\n")
```

```python
from tqdm import tqdm_notebook
```

```python
for phone in tqdm_notebook(phones):
    X_train = np.load("./Cambridge_bigger Phones Train/"+phone+"_X.npy",allow_pickle=True)
    Xmax = np.max(X_train,(0,1,2))
    Xmin = np.min(X_train,(0,1,2))
    np.save("Cambridge_bigger Train MinMax/"+"min_"+phone+".npy", Xmin)
    np.save("Cambridge_bigger Train MinMax/"+"max_"+phone+".npy", Xmax)
```

```python
# For Tanh work
for phone in tqdm_notebook(phones):
    X_train = np.load("./Cambridge_bigger Phones Train/"+phone+"_X.npy",allow_pickle=True)
    Xmax = np.max(np.abs(X_train),(0,1,2))
    np.save("Cambridge_bigger Train AbsMax/"+"max_"+phone+".npy", Xmax)
```
